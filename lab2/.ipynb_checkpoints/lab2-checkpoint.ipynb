Proučite i skicirajte model zadan objektom net u skripti train.py. Odredite veličine tenzora te broj parametara u svakom sloju. Odredite veličinu receptivnog polja značajki iz posljednjeg (drugog) konvolucijskog sloja.

Skica napravljena pomoću alata: https://math.mit.edu/ennui/

Broj parametara konvolucijskog sloja: broj filtera * dimenzija filtera + broj filtera (bias)

Slojevi s lijeva na desno:
- input (N * 1 * 28 * 28), 0 parametara
- konvolucijski sloj:
    - ulaz: N * 1 * 28 * 28
    - 1 kanal, 16 filtera dimenzije 5x5 -> 16 * 5 * 5 + 16 = 416 parametara
    - izlaz: N * 16 * 28 * 28
- pooling sloj:
    - ulaz: N * 16 * 28 * 28
    - nema parametara
    - izlaz: N * 16 * 14 * 14
- konvolucijski sloj:
    - ulaz: N * 16 * 14 * 14
    - 16 kanala, 32 filtera, dimenzije 5x5 -> 16 * 32 * 5 * 5 + 32 = 12832 parametara
    - izlaz: N * 32 * 14 * 14
- pooling sloj:
    - ulaz: N * 32 * 14 * 14
    - nema parametara
    - izlaz: N * 32 * 7 * 7
- flatten sloj:
    - ulaz: N * 32 * 7 * 7
    - nema parametara
    - izlaz: N * 1568
- potpuno povezani sloj:
    - ulaz: N * 1568
    - 512 neurona, svaki prima 1568 ulaza -> 512 * 1568 + 512 (bias) = 803328 parametara
    - izlaz: N * 512
- potpuno povezani sloj:
    - ulaz: N * 512
    - 10 neurona, svaki prima 512 ulaza -> 10 * 512 + 10 = 5130 parametara
    - izlaz: N * 10
- softmax izlaza (N * 10), 0 parametara

ukupno 416 + 12832 + 803328 + 5130 = 866506 parametara
